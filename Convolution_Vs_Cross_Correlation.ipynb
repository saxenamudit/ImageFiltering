```
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv_imshow
from skimage import io
from skimage import transform as ts
```

#Convolution

```
  url="https://upload.wikimedia.org/wikipedia/commons/d/d1/TUDelftSeal.png"  
  img = io.imread(url)
  cv_imshow(img)
```

```

plt.subplot(131),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
PSFkernel = np.array([[1,4,7,4,1],[4,16,26,16,4],[7,26,41,26,7],[4,16,26,16,4],[1,4,7,4,1]],np.float32)/273
outputGS = cv.filter2D(img,-1,PSFkernel)
plt.subplot(132),plt.imshow(outputGS),plt.title('Gaussian blur')
plt.xticks([]), plt.yticks([])
#another way to do the exact same is as follows
outputGSI = cv.GaussianBlur(img,(5,5),0)
plt.subplot(133),plt.imshow(outputGSI),plt.title('Gaussian blur (inbuilt)')
plt.xticks([]), plt.yticks([])
plt.show()
```

#Cross-Correlation

```

baseImage = io.imread("https://drive.google.com/uc?id=1cL0iXivYIOSGdG4JXQM-Gl5NutVODxzx")
findImage = io.imread("https://drive.google.com/uc?id=1q8ABcoxsmC4MNoQpAUEXx8ViXJRCRwFp")


plt.subplot(121),plt.imshow(baseImage),plt.title('Base image')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(findImage),plt.title('Find image')
plt.xticks([]), plt.yticks([])
plt.show()


result = cv.matchTemplate(baseImage, findImage, cv.TM_CCORR_NORMED)
plt.subplot(121),plt.imshow(baseImage),plt.title('Reference')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(result),plt.title('Result')
plt.xticks([]), plt.yticks([])
```
